```python
import numpy as np
```

```python
def process_data(fileName):
    """
        处理训练数据，得到词典及词频
    """
    fi = open(fileName)
    data = fi.readlines()
    voc = dict() # word: index
    for sentence in data:
        for word in sentence.split():
            voc[word] = len(voc)
#             voc[word] = voc.get(word, 0) + 1
    print("词典大小V={}".format(len(voc)))

    return voc
```

```python
voc = process_data('data.txt')
```

```python
# Hyper parameters
num_epoches = 10
learning_rate = 1e-3
embed_size = 50
```

```python
def train(fileName):
    fi = open(fileName)
    data = fi.readlines()
    train_data = [] # training dataset
    voc = dict() # word: index
    for sentence in data:
        words = sentence.split()
        for word in words:
            if word not in voc:
                voc[word] = len(voc)
        contexts = [voc[word] for word in words]
        targets = contexts[1:]
        
        instances = zip(contexts, targets)
        train_data.extend(instances)
        
    for key, value in voc.iteritems():
        print(key, value)

    W_ih = np.random.rand(len(voc), embed_size) # input-> hidden
    W_ho = np.random.rand(embed_size, len(voc)) # hidden-> output
    
    for epoch in range(num_epoches):
        np.random.shuffle(train_data)
        for (context, target) in train_data:
            h = W_ih[context]
            print(h.shape)
            
            
            
            
        
            
```
