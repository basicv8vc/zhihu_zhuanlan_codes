```python
import numpy as np
```

```python
# Hyper parameters
num_epoches = 10
learning_rate = 1e-3
embed_size = 50
```

```python
def softmax(v):
    v = v - np.max(v)
    exp_v = np.exp(v)
    return exp_v/np.sum(exp_v)
```

```python
def train(fileName):
    fi = open(fileName)
    data = fi.readlines()
    train_data = [] # training dataset
    voc = dict() # word: index
    for sentence in data:
        words = sentence.split()
        for word in words:
            if word not in voc:
                voc[word] = len(voc)
        contexts = [voc[word] for word in words]
        targets = contexts[1:]
        
        instances = zip(contexts, targets)
        train_data.extend(instances)
        

    V = len(voc) # vocabulary size
    W_ih = np.random.rand(len(voc), embed_size) # input-> hidden
    W_ho = np.random.rand(embed_size, len(voc)) # hidden-> output
    
    for epoch in range(num_epoches):
        np.random.shuffle(train_data)
        for (context, target) in train_data:
            label = np.zeros(V)
            label[target] = 1.
            label = label.reshape(V, 1) # (V, 1)
            # forward-pass
            h = W_ih[context].reshape(embed_size, 1) # (N, 1)
            u = np.dot(W_ho.T, h) # (V, 1)
            y = softmax(u)
            # backward-pass
            W_ho = W_ho - learning_rate*h*(y-label).T
            W_ih[context] = W_ih[context] - learning_rate*(y-label)*W_ho[]
            
            
            
            
            
            
            print(u.shape)
```
