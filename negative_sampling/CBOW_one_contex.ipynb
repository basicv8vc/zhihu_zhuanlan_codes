```python
def train(fileName):
    fi = open(fileName)
    data = fi.readlines()
    train_data = [] # training dataset
    voc = dict() # word: index
    for sentence in data:
        words = sentence.split()
        for word in words:
            if word not in voc:
                voc[word] = len(voc)
        contexts = [voc[word] for word in words]
        targets = contexts[1:]
        
        instances = zip(contexts, targets)
        train_data.extend(instances)
        
    for key, value in voc.iteritems():
        print(key, value)

    W_ih = np.random.rand(len(voc), embed_size) # input-> hidden
    W_ho = np.random.rand(embed_size, len(voc)) # hidden-> output
    
    for epoch in range(num_epoches):
        np.random.shuffle(train_data)
        for (context, target) in train_data:
            h = W_ih[context].reshape(embed_size, 1) # (emb_size, 1)
            u = np.dot(W_ho.T, h) # (V, 1)
            
            
            
            print(u.shape)
            
            
            
            
        
            
```
