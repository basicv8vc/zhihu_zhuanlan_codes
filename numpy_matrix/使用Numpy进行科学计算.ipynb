# 使用Numpy进行科学计算 

**ndarray**可以表示向量、矩阵、张量

```python
from __future__ import print_function, division, unicode_literals, absolute_import
import numpy as np
```

## 数量积

```python
a = np.random.rand(3)
a
```

```python
b = a + 1
b
```

## elementwise 计算

```python
a + b
```

```python
a - b
```

```python
a / b 
```

```python
a * b #  也被称为 Hadamard product
```

Hadamard product: $\bigodot$

## 点积/ dot product

在NN中非常常见的计算操作.

可以用向量内积、向量和矩阵、矩阵和矩阵计算

```python
np.dot(a, b)
```

```python
a.dot(b)
```

## 转置

```python
mat = np.random.rand(3, 5)
mat
```

```python
mat.T
```

```python
np.add(a, b)
```

```python
a + b
```

```python
np.dot(mat, mat.T)
```

```python
np.matmul(mat, mat.T)
```

## 向量化两个数组的element-wise product

$v_1=(a_{1},a_{2},...,a_{n})$, $v_2=(b_{1},b_{2},...,b_{n})$,

我们要得到

$ress=(a_{1} * b_{1}, a_{2}*b_{2},...,a_{n}*b_{n})$


第一种方式是for循环, 还可以把数组转为向量，进行计算。
